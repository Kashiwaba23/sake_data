```python
import requests
import pandas as pd
```

# Get Sake Brand Data

```python
brands_request = requests.get('https://muro.sakenowa.com/sakenowa-data/api/brands').json()
```

```python
b_list = brands_request['brands']
b_list
```

```python
brand_ids = [brand['id'] for brand in b_list]
names = [brand['name'] for brand in b_list]
brewery_ids = [brand['breweryId'] for brand in b_list]
```

```python
brand_df = pd.DataFrame({'brand_id':brand_ids,'brand_name':names, 'brewery_id':brewery_ids})
```

```python
brand_df.set_index('brand_id', inplace=True)
```

```python
brand_df
```

```python
brand_df.to_csv('brands.csv')
```

# Get Brewery Data

```python
brewery_response = requests.get('https://muro.sakenowa.com/sakenowa-data/api/breweries').json()['breweries']
```

```python
brewery_response
```

```python
brewery_ids = [brewery['id'] for brewery in brewery_response]
brewery_names = [brewery['name'] for brewery in brewery_response]
area_ids = [brewery['areaId'] for brewery in brewery_response]
```

```python
brewery_df = pd.DataFrame({'brewery_id':brewery_ids, 'brewery_name':brewery_names, 'area_id':area_ids})
```

```python
brewery_df.set_index('brewery_id', inplace=True)
```

```python
brewery_df
```

```python
brewery_df.to_csv('breweries.csv')
```

# Get Flavor Chart Data

```python
flavor_response = requests.get('https://muro.sakenowa.com/sakenowa-data/api/flavor-charts').json()
```

```python
flavor_response['flavorCharts']
```

```python
flavor_chart_df = pd.DataFrame.from_dict(flavor_response['flavorCharts'])
```

```python
flavor_chart_df.rename(columns={'f1':'華やか', 'f2':'芳醇', 'f3':'重厚', 'f4':'穏やか', 'f5':'ドライ', 'f6':'軽快'}, inplace=True)
```

```python
flavor_chart_df.set_index('brandId', inplace=True)
flavor_chart_df
```

```python
flavor_chart_df.to_csv('flavor_chart.csv')
```
